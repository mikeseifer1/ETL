```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
tables = pd.read_html(url)
tables
```

```python
sp500 = tables[0]
sp500
```

```python
sp500_new = sp500[["Symbol","GICS Sector","GICS Sub Industry","Headquarters Location"]]
sp500_new
```

```python
sp500_rename = sp500_new.rename(columns={"Symbol":"symbol","GICS Sector":"sector","GICS Sub Industry":"industry","Headquarters Location":"headquarters"})
sp500_rename                      
```

```python
csv_file_main = "ETL/sp-500-index-08-30-2020-main.csv"
sp_main = pd.read_csv(csv_file_main)
sp_main.tail()
```

```python
sp_main = sp_main[:-1]
sp_main.tail()
```

```python
sp_main_new = sp_main[["Symbol","Name","Last"]]
sp_main_new
```

```python
sp_main_new_rename = sp_main_new.rename(columns={"Symbol":"symbol","Last":"last_price","Name":"company_name"})
sp_main_new_rename
```

```python
csv_file_fundamental = "ETL/sp-500-index-08-30-2020-fundamental.csv"
sp_fundamental = pd.read_csv(csv_file_fundamental)
sp_fundamental.tail()
```

```python
sp_fundamental = sp_fundamental[:-1]
sp_fundamental.tail()
```

```python
sp_fundamental_new = sp_fundamental[["Symbol","Market Cap","P/E ttm","Div Yield"]]
sp_fundamental_new
```

```python
sp_fundamental_new_rename = sp_fundamental_new.rename(columns={"Symbol":"symbol","Market Cap":"market_cap","P/E ttm":"price_earnings_ratio","Div Yield":"dividend_yield"})
sp_fundamental_new_rename.dropna(subset=['market_cap'],inplace=True)
sp_fundamental_new_rename
```

```python
csv_file_performance = "ETL/sp-500-index-08-30-2020-performance.csv"
sp_performance = pd.read_csv(csv_file_performance)
sp_performance.tail()
```

```python
sp_performance = sp_performance[:-1]
sp_performance.tail()
```

```python
sp_performance_new = sp_performance[["Symbol","YTD %Chg","52W %Chg"]]
sp_performance_new
```

```python
sp_performance_new_rename = sp_performance_new.rename(columns={"Symbol":"symbol","YTD %Chg":"ytd_percentage_change","52W %Chg":"one_year_percentage_change"})
sp_performance_new_rename
```

```python
csv_file_technical = "ETL/sp-500-index-08-30-2020-Technical.csv"
sp_technical = pd.read_csv(csv_file_technical)
sp_technical.tail()
```

```python
sp_technical = sp_technical[:-1]
sp_technical.tail()
```

```python
sp_technical_new = sp_technical[["Symbol","52W Low","52W High"]]
sp_technical_new
```

```python
sp_technical_new_rename = sp_technical_new.rename(columns={"Symbol":"symbol","52W Low":"one_year_low","52W High":"one_year_high"})
sp_technical_new_rename
```

```python
connection_string = "postgres:postgres@localhost:5432/Stocks"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
sp500_rename.to_sql(name='sp_wiki', con=engine, if_exists='append', index=False)
```

```python
sp_main_new_rename.to_sql(name='sp_main', con=engine, if_exists='append', index=False)
```

```python
sp_fundamental_new_rename.to_sql(name='sp_fundamentals', con=engine, if_exists='append', index=False)
```

```python
sp_performance_new_rename.to_sql(name='sp_performance', con=engine, if_exists='append', index=False)
```

```python
sp_technical_new_rename.to_sql(name='sp_technical', con=engine, if_exists='append', index=False)
```
